```python
# For data manipulation and analysis
import pandas as pd
import numpy as np

# For data visualization
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8-darkgrid')
import seaborn as sns

# Set matplotlib to display visuals inline in a Jupyter Notebook
%matplotlib inline
```

```python
#loading climate data
cl_data = pd.read_csv("/Users/sandeepsawhney/Desktop/VSCode/MachineLearning/SimulatedClimateChangeProject/data/simulated_climate_data.csv")

#displays the first few pieces of data in the dataframe
print(cl_data.head())
```

```python
#prints unique pieces of data for the given categories (columns)
print(cl_data['Climate Condition'].unique())
```

```python
#visualizing change in temperature anomoly over the years (deviation from normal conditions)

plt.figure(figsize = (12,6))
plt.plot(cl_data['Year'], cl_data['Temperature Anomaly (°C)'], marker='o', linestyle='-', color='b')
plt.title('Global Temperature Anomoly Change over Time')
plt.xlabel('Year')
plt.ylabel('Deviation (°C)')
plt.show()
```

```python
import numpy as np
```

```python
timeBins = pd.cut(cl_data['Year'], bins=np.arange(1900, 2030, 10))
#gets the mean temperature anomaly by the periods in the bins
temp_by_time_period = cl_data.groupby(timeBins)['Temperature Anomaly (°C)'].mean()
```

```python
plt.figure(figsize=(12, 8))
temp_by_time_period.plot(marker='o', linestyle='-', color='skyblue')
plt.title('Average Temperature Anomaly by Time Period')
plt.xlabel('Time Period')
plt.ylabel('Temperature Anomaly (°C)')
plt.grid(True)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='Climate Condition', y='Temperature Anomaly (°C)', data=cl_data)
plt.title('Distribution of Temperature Anomalies by Climate Condition')
plt.xlabel('Climate Condition')
plt.ylabel('Temperature Anomaly (°C)')
plt.show()
```
